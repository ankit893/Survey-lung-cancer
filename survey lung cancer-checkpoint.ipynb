```python
# Summary Statistics
summary_stats = data.describe()

# Correlation Matrix
correlation_matrix = data.corr()
```

```python
# Plotting functions
def plot_age_distribution(data):
    plt.figure(figsize=(10, 6))
    sns.distplot(data['AGE'], bins=20, kde=True)
    plt.title('Age Distribution of Patients')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
```

```python
def plot_smoking_vs_lung_cancer(data):
    plt.figure(figsize=(10, 6))
    sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=data)
    plt.title('Smoking vs Lung Cancer')
    plt.xlabel('Smoking')
    plt.ylabel('Count')
    plt.show()
```

```python
def plot_gender_distribution(data):
    plt.figure(figsize=(10, 6))
    sns.countplot(x='GENDER', hue='LUNG_CANCER', data=data)
    plt.title('Gender Distribution and Lung Cancer')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.show()
```

```python
def plot_symptoms_analysis(data, symptom):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=symptom, hue='LUNG_CANCER', data=data)
    plt.title(f'{symptom.replace("_", " ").title()} and Lung Cancer')
    plt.xlabel(symptom.replace("_", " ").title())
    plt.ylabel('Count')
    plt.show()
```

```python
# Execute the plotting functions
plot_age_distribution(data)
plot_smoking_vs_lung_cancer(data)
plot_gender_distribution(data)
plot_symptoms_analysis(data, 'COUGHING')
plot_symptoms_analysis(data, 'SHORTNESS OF BREATH')

summary_stats, correlation_matrix
```

```python
#Heatmap of the Correlation Matrix
def plot_correlation_heatmap(data):
    plt.figure(figsize=(14, 10))
    sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()

plot_correlation_heatmap(data)
```

```python
#Age and Gender Distribution with Lung Cancer

def plot_age_gender_lung_cancer(data):
    plt.figure(figsize=(14, 6))
    sns.boxplot(x='GENDER', y='AGE', hue='LUNG_CANCER', data=data)
    plt.title('Age and Gender Distribution with Lung Cancer')
    plt.xlabel('Gender')
    plt.ylabel('Age')
    plt.show()

plot_age_gender_lung_cancer(data)
```

```python
#Box Plot of Age vs. Lung Cancer

def plot_age_vs_lung_cancer(data):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='LUNG_CANCER', y='AGE', data=data)
    plt.title('Box Plot of Age vs Lung Cancer')
    plt.xlabel('Lung Cancer')
    plt.ylabel('Age')
    plt.show()

plot_age_vs_lung_cancer(data)
```

```python
#Pair Plot

def plot_pair_plot(data):
    sns.pairplot(data, hue='LUNG_CANCER', diag_kind='kde')
    plt.title('Pair Plot of Features')
    plt.show()

plot_pair_plot(data)
```

```python
#Distribution of Symptoms by Lung Cancer Status

def plot_symptoms_distribution(data, symptoms):
    for symptom in symptoms:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=symptom, hue='LUNG_CANCER', data=data)
        plt.title(f'{symptom.replace("_", " ").title()} and Lung Cancer')
        plt.xlabel(symptom.replace("_", " ").title())
        plt.ylabel('Count')
        plt.show()

symptoms = ['COUGHING', 'SHORTNESS OF BREATH', 'CHEST PAIN', 'WHEEZING'] #, 'FATIGUE'
plot_symptoms_distribution(data, symptoms)
```

```python
#Heatmap of the Correlation Matrix

def plot_correlation_heatmap(data):
    plt.figure(figsize=(14, 10))
    sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()

plot_correlation_heatmap(data)
```

```python
def plot_age_vs_lung_cancer(data):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='LUNG_CANCER', y='AGE', data=data)
    plt.title('Box Plot of Age vs Lung Cancer')
    plt.xlabel('Lung Cancer')
    plt.ylabel('Age')
    plt.show()

plot_age_vs_lung_cancer(data)
```

```python
def plot_pair_plot(data):
    numeric_data = data.select_dtypes(include=['number'])
    numeric_data['LUNG_CANCER'] = data['LUNG_CANCER']
    sns.pairplot(numeric_data, hue='LUNG_CANCER', diag_kind='kde')
    plt.show()

plot_pair_plot(data)
```
